```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt


food_data = pd.read_csv('/content/drive/MyDrive/FoodBalanceSheets.csv',encoding='latin1')
food_data.head()
print(food_data.shape)
print(food_data.nunique())
print(food_data.head())
```

```python
# find the mean of each column with missing data 
mean_2014 = round(food_data['Y2014'].mean(), 3)
mean_2015 = round(food_data['Y2015'].mean(), 3)
mean_2016 = round(food_data['Y2016'].mean(), 3)
mean_2017 = round(food_data['Y2017'].mean(), 3)
mean_2018 = round(food_data['Y2018'].mean(), 3)
```

```python
# fill the missing column with respective mean
food_data[['Y2014']] = food_data[['Y2014']].fillna(mean_2014)
food_data[['Y2014']].isnull().sum()
food_data[['Y2015']] = food_data[['Y2015']].fillna(mean_2015)
food_data[['Y2015']].isnull().sum()
food_data[['Y2016']] = food_data[['Y2016']].fillna(mean_2015)
food_data[['Y2016']].isnull().sum()
food_data[['Y2017']] = food_data[['Y2017']].fillna(mean_2015)
food_data[['Y2017']].isnull().sum()
food_data[['Y2018']] = food_data[['Y2018']].fillna(mean_2015)
food_data[['Y2018']].isnull().sum()
```
